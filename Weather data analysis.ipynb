## Mean temp record per month

```python
df.head()
```

```python
df.loc[df.index.month == 1, 'Temp (C)'].mean()
```

```python
mean_temp = {}

for month in range(1, 13):
    mean_temp[month] = df.loc[df.index.month == month, 'Temp (C)'].mean()
    
pd.Series(mean_temp)    
```

```python
# another solution
# using pivot 

mean_temp_df = df.pivot_table(values = 'Temp (C)', index = df.index.month, aggfunc = np.mean)
mean_temp_df
```

## Extract month name from date and store it in a column

```python
df['Month'] = df.index.month
df['Month'] = df['Month'].apply(lambda x: calendar.month_name[x])
df.head()
```

## Get The list of months in which it rained with thunderstrom

```python
df.head()
```

```python
df['Weather'].unique()
```

```python
def checklist(x):
    weatherlist = ['rain', 'thunderstorms']
    count = 0
    
    for i in weatherlist:
        if i in x:
            count += 1;
    if(count == 2):
        return True
    else:
        return False
    
print("2) Get the list of months in which it rained with thunderstrom\n")
rained = df['Weather'].str.lower().apply(checklist)
# rained
month_list = df[rained]['Month'].unique()
print(month_list)    
```
